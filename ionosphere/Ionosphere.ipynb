```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd


data = pd.read_csv('ionosphere.data')
X = data.iloc[:, :-1].to_numpy()
y = data.iloc[:, -1].to_numpy()


        
```

```python
print(type(X), type(y))
print(X.shape, y.shape)
print(np.round(X, 1)[:2])
print(y[:22])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=14)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=1, n_jobs=-1, leaf_size=10)
print(knn, '\n')
knn.fit(X_train, y_train)


```

```python
y_pred = knn.predict(X_test)
print(np.round(100*np.mean(y_pred == y_test), 1), "%")
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn, X_train, y_train, scoring = 'accuracy', cv=4)
#print(scores)
print(np.round(100*np.mean(scores), 1), "%")

print("Разница между прогнозами на тестовой выборке и валидации:", 
      np.round(100*np.mean(y_pred == y_test) - 100*np.mean(scores), 1), "%")
```

```python

scores_all = cross_val_score(knn, X, y, scoring = 'accuracy', cv=3)
print(np.round(100*np.mean(scores_all), 1))
```
